```python
# THEORETICAL BACKGROUND
import random
import matplotlib.pyplot as plt

mu = 0  # Mean of the distribution
sigma = 1  # Standard deviation
num_samples = 100000  # Number of random numbers to generate

random_numbers = [random.gauss(mu, sigma) for _ in range(num_samples)]

plt.hist(random_numbers, bins=300, density=True, alpha=0.6, color='b')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Gaussian Distribution')
plt.grid(True)

plt.show()
```

```python
from skimage import io, color

# Load the image
image = io.imread("donald_trump_gauss.jpg")

# Transform it to a grayscale image
gray_image = color.rgb2gray(image)

# Obtain the dimensions
height, width = gray_image.shape

# Loop through all the pixels of the image
for x in range(height):
    for y in range(width):
        gauss_value = random.gauss(gray_image[x, y], 0.3) # Sigma = 0.3 (Standard deviation)
        gray_image[x, y] = gauss_value

# Plot the gaussian image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```
